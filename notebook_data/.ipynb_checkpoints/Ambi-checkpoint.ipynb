```python
import pandas as pd
df = pd.read_csv('Ambi3_data.csv')

df
```

```python
def find_text(temp_df, text,case=False):
    location = temp_df[temp_df['TEXT'].str.contains(text,case=case)].index.tolist()
    return location

def find_next_word(temp_df, text, nearest_text, case=False):
    pos = find_text(temp_df,text,case=case)
    pos = [item +1 for item in pos]
#     print(pos)
    new_pos = list()
    for item in pos:
#         temp = df.loc[]
        temp_value= temp_df.loc[item]['TEXT']
#         print(temp_value.lower())
        if temp_value.lower().__contains__(nearest_text.lower()):
            new_pos.append(item-1)
        
    return new_pos
```

```python
pos_item = find_text(df,'QTY',case=True)
# df.loc[pos_item[0]:pos_total[0]]
# pos_total = find_next_word(df, '', 'and')
table_data = df.loc[pos_item[0]:]
for index , row in table_data.iterrows():
    from pprint import pprint
prodMessage = {}
table_row = []
for index, row in table_data.iterrows():
    if 50 < row['X1'] < 100:
        if ('price' in prodMessage) :
            table_row.append(prodMessage)
            prodMessage = {}
        prodMessage['Sr'] = row['TEXT']     
        prodMessage['product'] = ''
    elif ('Sr' in prodMessage) and (120 < row['X1'] < 300):
        prodMessage['product'] =prodMessage['product']+" "+row['TEXT']
    elif ('product' in prodMessage) and (400 < row['X1'] < 420):
        prodMessage['qty'] = row['TEXT']
    elif ('qty' in prodMessage) and (490 < row['X2'] < 500):
        prodMessage['price'] = row['TEXT']

table_row.append(prodMessage)
pprint(table_row)
# table_data
```
